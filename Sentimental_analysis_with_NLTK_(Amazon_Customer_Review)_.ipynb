```python
import pandas as pd
import random

# Sample data
data = {
    "review_id": [f"R{i:08d}" for i in range(1, 101)],
    "customer_id": [f"C{i:08d}" for i in range(1, 101)],
    "product_id": [f"P{random.randint(1000, 9999)}" for _ in range(100)],
    "review_date": pd.date_range(start='2022-01-01', periods=100).tolist(),
    "review_text": [
        "Great product, highly recommend!" if i % 5 == 0 else
        "The product quality was not as expected." if i % 5 == 1 else
        "Fast delivery, good packaging." if i % 5 == 2 else
        "Received a damaged item, not happy." if i % 5 == 3 else
        "Value for money, satisfied with the purchase."
        for i in range(100)
    ],
    "star_rating": [random.randint(1, 5) for _ in range(100)],
}

# Create DataFrame
df = pd.DataFrame(data)

# Add review label based on star rating
df['review_label'] = df['star_rating'].apply(lambda x: 'positive' if x >= 4 else 'negative')

# Display the first few rows of the DataFrame
df=df[["review_label","review_text"]]
df
```

```python
df['Scores']=df['review_text'].apply(lambda review:sid.polarity_scores(review))
df
```

```python
df['Compund']=df['Scores'].apply(lambda d:d['compound'])
df
```

```python
df['Coumpound_score']=df['Compund'].apply(lambda score: 'positive' if score>=0 else 'negative')
df
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(df['review_label'],df['Coumpound_score'])
```

```python
classification_report(df['review_label'],df['Coumpound_score'])
```

```python
confusion_matrix(df['review_label'],df['Coumpound_score'])
```
